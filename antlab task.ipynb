```python
import PyPDF2
import pandas as pd
import requests
import re
#first of all i have write a function to download pdf from given url.by giving name of function as download_pdf
#step1:-
def download_pdf(pdf_url):
    response=requests.get(pdf_url)
    with open('temp_pdf.pdf', 'wb') as file:
        file.write(response.content)
    return 'temp_pdf.pdf'
#in this i have writen function to extract text on basis of page number by using libray called pypdf2
#step2:-
def extract_text_from_pdf(pdf_path, page_number):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        if page_number > num_pages:
            raise ValueError(f"Page number{page_number} is greater than the total number of pages.")
        page = reader.pages[page_number-1]  
        return page.extract_text()

#unction to take text  and extract attribute value from pdf below are attribute taken like brand,regulator,input voltage
#step3
def parse_text(text):
    attributes = {}
    brand_pattern=re.compile(r'Brand:\s*(.*)',re.IGNORECASE)
    brand_match=brand_pattern.search(text)
    if brand_match:
        attributes['Brand']=brand_match.group(1).strip()
    

    regulator_pattern=re.compile(r'Regulator\s*Function:\s*(.*(?:\n\s+.*)*)',re.IGNORECASE)
    regulator_match=regulator_pattern.search(text)
    if regulator_match:
        attributes['Regulator Function']=regulator_match.group(1).strip()
    

    input_voltage_pattern=re.compile(r'Input\s*Voltage\s*Range:\s*(.*(?:\n\s+.*)*)', re.IGNORECASE)
    input_voltage_match=input_voltage_pattern.search(text)
    if input_voltage_match:
        attributes['Input Voltage Range']=input_voltage_match.group(1).strip()
    

    
    return attributes
#mpn_data in this all the given pdf links and part numbers are given
#step4:-
mpn_data={
    'LM46000QPWPTQ1': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm46000-q1.pdf', 'page_number': 46},
    'LM46001QPWPTQ1': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm46001-q1.pdf', 'page_number': 49},
    'LM46002QPWPTQ1': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm46002-q1.pdf', 'page_number': 48},
    'LM5165XDRCT': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm5165.pdf', 'page_number': 3},
    'LM53600AQDSXTQ1': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm53600-q1.pdf', 'page_number': 3},
    'LM74610QDGKTQ1': {'pdf_link': 'http://www.ti.com/lit/ds/symlink/lm74610-q1.pdf', 'page_number': 25}
}

#here for loop is used to iterate through all like text etc and coverted to data fframe.
data_rows=[]
for mpn, info in mpn_data.items():
    pdf_path=download_pdf(info['pdf_link'])
    pdf_text=extract_text_from_pdf(pdf_path, info['page_number'])
    print(f"Extracted Text for {mpn} (Page {info['page_number']}):")
    print(pdf_text)  # Print the extracted text for debugging purposes
    attributes=parse_text(pdf_text)
    data_rows.append({'Manufacturer Part Number': mpn, **attributes})

df = pd.DataFrame(data_rows)
print(df)
df = pd.DataFrame(data_rows)
print(df)
```
